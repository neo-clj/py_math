# 91-94: Verifying Zeros Using a Graphing Device

## 91. $$ x^{3} - 3x^{2} - 4x + 12 = 0; [-4, 4] by [-15, 15] $$

```python
x = np.linspace(-5, 5, 10000)
y = (x**3) - (3 * x**2) - (4 * x) + 12

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

plt.plot(x, y)

plt.ylim(-15, 15)
plt.xlim(-4, 4)
plt.grid(True)
```
