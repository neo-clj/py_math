# Example 9.a
## $$ P_{1}(x) = x^{4} + x^{3} - 16x^{2} - 4x + 48 $$

```python
x = np.linspace(-5, 5, 10000)
y = (x**4) + (x**3) - (16*x**2) - (4*x) + 48

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

plt.plot(x, y)

plt.ylim(-100, 100)
plt.xlim(-5, 4)
plt.grid(True)
```

# 15-30 Graphing Factored Polynomials  
## Sketch the graph of the polynomial function. Make sure your graph shows all intercepts and exhibits the proper end behavior.

## 15. $$ P(x) = (x - 1)(x + 2) $$

```python
xs1 = np.linspace(-10, 10, 1000000)
ys1 = (xs1 - 1) * (xs1 + 2)

xs2 = np.array([-4, -3, -2, -1, -0.5, 0, 1, 2])
ys2 = (xs2 - 1) * (xs2 + 2)
   

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')


plt.plot(xs1, ys1)
plt.scatter(xs2, ys2)


plt.ylim(-10, 10)
plt.xlim(-10, 10)
plt.grid(True)
```

# Exercises 73-78: Families of Polynomials  
## Graph the family of polynomials in the same viewing rectangle, using the given values of c. Explain how changing the value of c affects the graph.

# Exercise 73:
## $$ P(x) = cx^{3}; c = 1, 2, 5, \frac{1}{2} $$

```python
xs = np.linspace(-10, 10, 1000000)
yss = []
cs = [1, 2, 5, 1/2]
strings = []


for c in cs:
    ys = c * xs**3
    yss.append(ys)
    string = '{0} * x**3'.format(c)
    string = sympy.sympify(string)
    strings.append(string)
    

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

for ys, string in zip(yss, strings):
    plt.plot(x, ys, label=string)

plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
               ncol=2, mode="expand", borderaxespad=0.)

plt.ylim(-100, 100)
plt.xlim(-10, 10)
plt.grid(True)
```
