```python
import numpy as np
import sympy
```

```python
import matplotlib.pyplot as plt
```

```python
# figure 2, p.332
```

```python
bases = np.array([1/2, 1/3, 1/5, 1/10, 2, 3, 5, 10])
xs = np.linspace(-10, 10, 1000000)
yss = []
strings = []

for base in bases:
    ys = base ** xs
    yss.append(ys)
    string = '{0} ** x'.format(base)
    string = sympy.sympify(string)
    strings.append(string)
    
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

for ys, string in zip(yss, strings):
    plt.plot(xs, ys, label=string)
    
plt.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
               ncol=2, mode="expand", borderaxespad=0.)

plt.ylim(-1, 3)
plt.xlim(-3, 3)
plt.grid(True)
```
