```python
x1 = np.linspace(-1000, 1000, 10000)
y1 = x1 ** 2

x2 = np.linspace(-1000, 1000, 10000)
y2 = 3 * x2 + 10 



fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

plt.plot(x1, y1, color='b')
plt.plot(x2, y2, color='r')

plt.ylim(-2, 60)
plt.xlim(-15, 15)
plt.grid(True)
```
