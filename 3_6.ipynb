```python
x1 = np.linspace(-1000, 2.9, 10000)
y1 = (x1**2 - 4*x1 - 5) / (x1 - 3)
#y1 = x1 - 1 - (8 / (x1 - 3))

x2 = np.linspace(3.1, 1000, 10000)
y2 = (x2**2 - 4*x2 - 5) / (x2 - 3)
#y2 = x2 - 1 - (8 / (x2 - 3))

x3 = np.linspace(-1000, 1000, 10000)
y3 = x3 - 1

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

plt.plot(x1, y1)
plt.plot(x2, y2)
plt.plot(x3, y3, linestyle='--')
plt.vlines(x=3, ymin=-20, ymax=20, linestyles='--',)

plt.ylim(-20, 20)
plt.xlim(-15, 15)
plt.grid(True)
```

# 69. $$ r(x) = \frac{x^{2}}{x - 2} $$ 

```python
x1 = np.linspace(-1000, 1.9, 10000)
y1 = x1 + 2 + (4 / (x1 - 2))

x2 = np.linspace(2.1, 1000, 10000)
y2 = x2 + 2 + (4 / (x2 - 2))

x3 = np.linspace(-1000, 1000, 10000)
y3 = x3 + 2

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

plt.plot(x1, y1, color='b')
plt.plot(x2, y2, color='b')
plt.plot(x3, y3, linestyle='--', color='r')
plt.vlines(x=2, ymin=-20, ymax=20, linestyles='--', color='r')

plt.ylim(-20, 20)
plt.xlim(-15, 15)
plt.grid(True)
```
