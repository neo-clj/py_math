$$ y = \tan x $$ 

```python
xs = []
ys = []
asymptotes = []

for i in range(-2, 3):
    x = np.linspace((i * np.pi) + (-np.pi / 2) + 0.000001, 
                    (i * np.pi) + (np.pi / 2) - 0.000001, 10000)
    y = np.tan(x)
    asymptote = (i * np.pi) + (np.pi / 2)
    
    xs.append(x)
    ys.append(y)
    asymptotes.append(asymptote)
    
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')

for x, y, asymptote in zip(xs, ys, asymptotes):
    plt.plot(x, y, color='b')
    plt.vlines(x=asymptote, ymin=-20, ymax=20, linestyles='--', color='r')

plt.ylim(-12, 12)
plt.xlim(-12, 12)
plt.grid(True)

```
