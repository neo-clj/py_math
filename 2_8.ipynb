# 103. Find the inverse of the following functions

## (a) $$ f(x) = \frac{2x + 1}{5} $$ 

```python
x = np.linspace(-10, 10, 100000)
y = (2*x + 1) / 5

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('zero')

plt.plot(x, y)
plt.plot(y, x)
plt.plot(x, x)

plt.ylim(-10, 10)
plt.xlim(-10, 10)
plt.grid(True)
```

## (b) $$ f(x) = \frac{1}{x} $$

```python
x1 = np.linspace(0.000001, 30, 10000)
y1 = (1/x1)

x2 = np.linspace(-0.000001, -30, 10000)
y2 = (1/x2)

fig, axes = plt.subplots(2, 1, sharex=True, sharey=True)



axes[0].plot(x1, y1)
axes[0].plot(x2, y2)

axes[1].plot(x1, y1)
axes[1].plot(x2, y2)

plt.ylim(-5, 5)
plt.xlim(-5, 5)

axes[0].grid(True)
axes[0].spines['left'].set_position('center')
axes[0].spines['bottom'].set_position('zero')

axes[1].grid(True)
axes[1].spines['left'].set_position('center')
axes[1].spines['bottom'].set_position('zero')
```

# (b) $$ f(x) = 3 - \frac{1}{x} $$

```python
x1 = np.linspace(0.000001, 30, 10000)
y1 = 3 - (1/x1)

x2 = np.linspace(-0.000001, -30, 10000)
y2 = 3 - (1/x2)

fig, axes = plt.subplots(2, 1, sharex=True, sharey=True)



axes[0].plot(x1, y1)
axes[1].plot(x2, y2)

axes[0].plot(y1, x1)
axes[1].plot(y2, x2)

plt.ylim(-10, 10)
plt.xlim(-10, 10)

axes[0].grid(True)
axes[0].spines['left'].set_position('center')
axes[0].spines['bottom'].set_position('zero')

axes[1].grid(True)
axes[1].spines['left'].set_position('center')
axes[1].spines['bottom'].set_position('zero')
```
