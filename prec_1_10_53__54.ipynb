```python
def convert_input(string):
    expr = sympify(string, evaluate=False)
    sym, = expr.free_symbols
    x = Symbol(sym.name, real=True)
    expr = expr.subs(sym, x, evaluate=False)
    return expr, x
```

```python
def convert_to_frac(string):
    expr = sympify(string, evaluate=False)
    sym, = expr.free_symbols
    x = Symbol(sym.name, real=True)
    expr = expr.subs(sym, x)
    frac = cancel(together(expr))
    return frac, x
```

```python
def plot_curve(expr, x, x_min, x_max, y_min, y_max):
    """Plot y=expr(x) over the specified domain."""
    func = lambdify([x], expr)
    xs = np.linspace(x_min, x_max, 200)
    plt.plot(xs, func(xs))
    plt.ylim(y_min, y_max)
    plt.xlim(x_min, x_max)
    plt.grid(True)
```

```python
for i in [-2, -1,0, 1, 2]:
    expr, x = convert_to_frac('1/2 * x + {0}'.format(i))
    pprint(expr)
    print('-'*60)
```

```python
for b in [0, -1, 1, -3, 3, -6, 6]:
    expr, x = convert_input('-2 * x + {0}'.format(b))
    plot_curve(expr, x, -6, 6, -6, 6)
```

```python
for m in [0, -0.25, 0.25, -0.75, 0.75, -1.5, 1.5]:
    expr, x = convert_input('{0} * x - 3'.format(m))
    pprint(expr)
    print('-'*60)
```

```python
for m in [0, -0.25, 0.25, -0.75, 0.75, -1.5, 1.5]:
    if m == 0:
        m += 0.0000000000000001 # because 0.0 raise an error
    expr, x = convert_input('{0} * x - 3'.format(m))
    plot_curve(expr, x, -5, 5, -5, 5)
```
